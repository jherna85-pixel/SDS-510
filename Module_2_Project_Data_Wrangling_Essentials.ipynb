Create a list of article information that contain the key words (this API is not able to pull articles based on an ID, but we can get all of the information when we use get_everything)



1.   Find range of last 3 days, and format timestamp for the request (YYYY, MM, DD)
2.   Use 'get_everything' to look through all articles and in each look for the keywords: cheese, pickles, buns, meat, sesame. Create a list of urls of these found articles with matches
3. Go through the list of articles and get the following information: source, title, author, description, url, and keyword





```python
time_now = datetime.datetime.now(datetime.timezone.utc)
time_72 = time_now - datetime.timedelta(hours=72)
from_date = time_72.strftime('%Y-%m-%d')
to_date = time_now.strftime('%Y-%m-%d')

article_list = []
keywords = ['cheese','pickles','buns','meat','sesame']
for word in keywords:
    sleep(2)
    results = api.get_everything(
        q=word,
        from_param=from_date,
        to=to_date,
        language='en',
        sort_by='publishedAt',
        page_size=100)
    for eacharticle in results.get('articles', []):
        pub_time = datetime.datetime.fromisoformat(eacharticle['publishedAt'].replace('Z','+00:00'))
        if pub_time >= time_72:
            article_list.append({
                'source': eacharticle['source']['name'],
                'title': eacharticle['title'],
                'author': eacharticle['author'],
                'description': eacharticle['description'],
                'url': eacharticle['url'],
                'keyword_match': word})
```

Clean up text and output as a CSV


1.   Create news_hamburglar.csv
2.   Clean text, replace commas and new lines with spaces




```python
with open('news_hamburglar.csv', 'w') as nhcsv:

    nhcsv.write('source, title, author, description, url, keyword_match\n')


    for eacharticle in article_list:

        clean_title = str(eacharticle['title']).replace(',', ' ').replace('\n', ' ')
        clean_description = str(eacharticle['description']).replace(',', ' ').replace('\n',' ')
        clean_author = str(eacharticle['author']).replace(',', ' ')

        line = (
            eacharticle['source'] + ', \'' +
            clean_title + '\', ' +
            clean_author + ', \'' +
            clean_description + '\', ' +
            eacharticle['url'] + ', \'' +
            eacharticle['keyword_match'] + '\n')

        nhcsv.write(line)
```
